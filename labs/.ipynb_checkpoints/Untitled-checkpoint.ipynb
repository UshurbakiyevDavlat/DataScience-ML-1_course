```python
 ctr = 0 
def Hanoi(num,froms,to,buf):
    if(num == 0):return 0
    global ctr
    ctr = ctr+1
    Hanoi(num-1,froms,buf,to)
    print("Disk num:"+ (str)(num) +" moved from "+ (str)(froms)+" to->"+ (str)(to))
    Hanoi(num-1,buf,to,froms)

numDisk = 8
froms = 1
to = 2
buf = 3


Hanoi(numDisk,froms,to,buf)
print("Number of operations:"+(str)(ctr))
```
