```python
# Import the necessary packages and modules
import matplotlib.pyplot as plt
import numpy as np

# Prepare the data
data = np.linspace(0, 9, 100)

# Plot the data
plt.plot(data, data, label='linear')
# Add label text
plt.ylabel('some numbers')
plt.xlabel('some numbers')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

<h3>TASK 2: Creating graphs </h3>

<b>Task 2.1</b> Plotting bar charts
<br/>
On the example below, the bar chart for Seasons was plotted the documentation [here](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.bar.html)

```python
plt.bar(data.Season.unique(), data.groupby('Season')['Sales'].sum(), width = 0.9, 
        color='#64fd00', ec='#414142')
plt.xlabel('Seasons')
plt.ylabel('Sales amount')
plt.title('Sales over seasons')
plt.show()
```

<b>Now, plot a bar char for Days and Revenue means </b>
<br/>
And write the summary below

```python
# Your code goes here
plt.bar(data.Day.unique(),data.groupby('Day')['Revenue'].mean(),width =0.9,color='#64fd00', ec='#414142')
plt.xticks(rotation=25)
plt.xlabel('Days')
plt.ylabel('Revenue amount')
plt.title('Revenue over Days')
plt.show()
```

<b>Task 2.3</b> Plotting bar charts
<br/> Plot two bar charts visualizing <b>Any month days + Sales </b> and <b> June days + Sales</b>
Use subplot, both graphs should be on the same figue
<br/>
And write the summary below

```python
# Your code goes here


fig,axs = plt.subplots(1,2, figsize=(10,4), sharey=True)

dataAll = data[data.Date.dt.month >= 1]
axs[0].bar(dataAll.Date.unique(), dataAll.groupby('Date')['Sales'].sum(), width = 0.4,
color='yellow', ec='black')
plt.xticks(rotation=45)

dataJune = data[data.Date.dt.month == 6]
axs[1].bar(dataJune.Date.unique(), dataJune.groupby('Date')['Sales'].sum(), width = 0.4,
color='orange', ec='black')
plt.xticks(rotation=45)


```

<b>TASK 3</b> Scatterplots

Take a look at well known <u>Temperature - Sales relationship</u>, visualized by scatterplot


```python
plt.scatter(data.Temperature, data.Sales, 50, marker = '*', color = '#2Ffd50')
plt.yticks(np.arange(0, 60, 5))
plt.xticks(np.arange(10, 130, 10))
plt.xlabel('Temperature')
plt.ylabel('Sales')
plt.axis('equal')
plt.title('Sales and Temperature relationship')
plt.show()
```

<b>TASK 5</b> Pie chart

<b>Task 5.1</b> Create a pie chart for your dataset. Use anymonth's sales as part of the whole
<br/>
And write the summary below

```python
# Your code goes here
ax1=plt.subplot()

ax1.pie(
        
       data.groupby('Day')['Sales'].sum(),
        
        labels  = data.Day.unique(),
 
    
    autopct = '%.1f',
    shadow  = True,
    
)

ax1.axis('equal')
plt.show()
```
