<b> Task 3 </b> Handle the missing data. Provide an explanation to why did you use particular
method.

```python
i = 0
sum = 0
while i < data.shape[1]:  
 if (data[str(data.iloc[:, i].name)].isnull().any()):   
  print('column -',str(data.iloc[:, i].name),' - have:', data[str(data.iloc[:, i].name)].isnull().sum(), 'NaN Data')
 sum += data[str(data.iloc[:, i].name)].isna().sum()
 i += 1 
print('Total NaNs -', sum)
```

```python
i = 0
sum = 0
while i < data.shape[1]:  
 if (data[str(data.iloc[:, i].name)].isnull().any()):   
  percent_missing = data[str(data.iloc[:, i].name)].isna().sum() * 100 / len(data)
  print('column -',str(data.iloc[:, i].name),' - have:', percent_missing, '%')
 i += 1
print(len(data))
```

```python
i = 0
sum = 0
while i < data.shape[1]:  
 if (data[str(data.iloc[:, i].name)].isnull().any()):   
  data.drop(data[data[str(data.iloc[:, i].name)].isnull()].index, axis=0, inplace = True)  #Dropping row - that have less than 5% of NaNs.
 i += 1
print(len(data))
```

```python
data.head(10)
```

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
i = 0
j = 0
k = 0
s = 0
print('less than min - outliers')
while i < data.shape[1]:  
 if (data.iloc[:, i].dtype == 'int64' or data.iloc[:, i].dtype == 'float64'):
  dataTrueLess = data[str(data.iloc[:, i].name)] < (Q1[str(data.iloc[:, i].name)] - (1.5 * IQR[str(data.iloc[:, i].name)]))
  while k < len(dataTrueLess):
    if (dataTrueLess[k]): 
      print(str(data.iloc[:, i].name), 'index -',k,', value -',  data[str(data.iloc[:, i].name)][k])
    k += 1
 k = 0
 i += 1 
print('----------------------------------------------------')   
print('more than max - outliers')
while j < data.shape[1]:  
 if (data.iloc[:, j].dtype == 'int64' or data.iloc[:, j].dtype == 'float64'): 
  dataTrueMore = data[str(data.iloc[:, j].name)] > (Q3[str(data.iloc[:, j].name)] + (1.5 * IQR[str(data.iloc[:, j].name)]))
  while s < len(dataTrueMore):
    if (dataTrueMore[s]): 
      print(str(data.iloc[:, j].name), 'index -',s,', value -',  data[str(data.iloc[:, j].name)][s])
    s += 1 
 s = 0
 j += 1 
```

<b> Task 5 </b> Visualize and analyze the relationships. Are there some strong patterns? Are
there any linear relationships? Include all findings to your summary. –

```python
plt.hist(data['CO(GT)'], density=False, bins=45)  
plt.ylabel('Count')
plt.xlabel('CO(GT)');
```

```python
plt.scatter(data['CO(GT)'], data['PT08.S1(CO)'], 5, marker = '^', color = '#2Ffd50')
plt.xlabel('CO(GT)')
plt.ylabel('PT08.S1(CO)')
plt.title('CO(GT) and PT08.S1(CO) Relation')
plt.show()
```

```python
plt.plot(data['C6H6(GT)'], data['PT08.S2(NMHC)'], label='linear Relation')
plt.ylabel('C6H6(GT)')
plt.xlabel('PT08.S2(NMHC)')
plt.legend()
plt.show()
```

```python
dataVisual = data.corr()
mask = np.triu(np.ones_like(dataVisual, dtype=bool))
f, ax = plt.subplots(figsize=(12, 8))
cmap = sns.diverging_palette(110, 10, as_cmap=True)
sns.heatmap(dataVisual, mask=mask, cmap=cmap, vmax=.99, center=0, square=True, linewidths=.99, cbar_kws={"shrink": .99})
```
