```python
dataVisual = data.corr()
mask = np.triu(np.ones_like(dataVisual, dtype=bool))
f, ax = plt.subplots(figsize=(12, 8))
cmap = sns.diverging_palette(110, 10, as_cmap=True)
sns.heatmap(dataVisual, mask=mask, cmap=cmap, vmax=.99, center=0, square=True, linewidths=.99, cbar_kws={"shrink": .99})
```

```python
data['height'].fillna(data['height'].mean(), inplace = True)
for j in data['weight'].index:
 if(np.isnan(data['weight'][j])):
  Height = data[data.height > data['height'].mean() - 0.2]
  TrueHeight = Height[Height.height < data['height'].mean() + 0.2]
data['weight'].fillna(TrueHeight['height'].mean(), inplace = True)
```

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
i = 0
j = 0
k = 0
s = 0
print('less than min - outliers')
while i < data.shape[1]:  
 if (data.iloc[:, i].dtype == 'int64' or data.iloc[:, i].dtype == 'float64'):
  dataTrueLess = data[str(data.iloc[:, i].name)] < (Q1[str(data.iloc[:, i].name)] - (1.5 * IQR[str(data.iloc[:, i].name)]))
  for k in dataTrueLess.index:
    if (dataTrueLess[k]): 
      print(str(data.iloc[:, i].name), 'index -',k,', value -',  data[str(data.iloc[:, i].name)][k])
 i += 1 
print('----------------------------------------------------')   
print('more than max - outliers')
while j < data.shape[1]:  
 if (data.iloc[:, j].dtype == 'int64' or data.iloc[:, j].dtype == 'float64'): 
  dataTrueMore = data[str(data.iloc[:, j].name)] > (Q3[str(data.iloc[:, j].name)] + (1.5 * IQR[str(data.iloc[:, j].name)]))
  for s in dataTrueMore.index:
    if (dataTrueMore[s]): 
      print(str(data.iloc[:, j].name), 'index -',s,', value -',  data[str(data.iloc[:, j].name)][s])
 j += 1 
```
