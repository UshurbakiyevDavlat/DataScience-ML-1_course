```python
# Here i am catching  the people whose weight is less than 40 killogram
# Task 4
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
i = 0
j = 0
print('Less than minimum - outlier')
while i < data.shape[1]:  
 if (data.iloc[:, i].dtype == 'int64' or data.iloc[:, i].dtype == 'float64'): 
    out=(data[str(data.iloc[:, i].name)] < (Q1[str(data.iloc[:, i].name)] - (1.5 * IQR[str(data.iloc[:, i].name)])))
    print(out)
 i += 1 

    
print('----------------------------------------------------')   
print('more than maximum - outliers')
while j < data.shape[1]:  
 if (data.iloc[:, j].dtype == 'int64' or data.iloc[:, j].dtype == 'float64'): 
    out = (data[str(data.iloc[:, j].name)] > (Q3[str(data.iloc[:, j].name)] + (1.5 * IQR[str(data.iloc[:, j].name)])))
    print(out)
 j += 1 


data.boxplot('weight')
data[data['weight']<40]  
```

```python
#Visualisation Task 5
plt.hist(data['height'], density=False, bins=45)  
plt.ylabel('Count')
plt.xlabel('height');

# Here we are showing counting of height values
```

```python
plt.scatter(data['height'], data['weight'], 5, marker = '^', color = '#2Ffd50')
plt.xlabel('height')
plt.ylabel('weight')
plt.title('height and weight Relation')
plt.show()
# Here we are showing relationship of height and weight values with scatter plot
```

```python
plt.plot(data['height'], data['weight'], label='linear Relation')
plt.ylabel('height')
plt.xlabel('weight')
plt.legend()
plt.show()
# there is  linear relation
```
